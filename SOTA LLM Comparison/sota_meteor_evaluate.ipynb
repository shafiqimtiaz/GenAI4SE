````python
from jury import Jury
import pandas as pd
import numpy as np

# Define the systems
systems = ["Mixtral", "ChatGPT", "Gemini", "Claude", "Microsoft Copilot"]

# Define the sets
sets = {
    "set1": {
        "predictions2": id1exp2,  # replace with your actual data
        "predictions4": id1exp4,  # replace with your actual data
        "references": id1ref  # replace with your actual data
    },
    "set2": {
        "predictions2": id2exp2,  # replace with your actual data
        "predictions4": id2exp4,  # replace with your actual data
        "references": id2ref  # replace with your actual data
    }
}

# Define the metrics
MT_METRICS = ["meteor"]
RUN_CONCURRENT = True  # set False to disable concurrency.

# Initialize the scorer
mt_scorer = Jury(metrics=MT_METRICS, run_concurrent=RUN_CONCURRENT)

# Prepare a dictionary to store the scores
scores_dict = {"System": [], "Experiment": [], "Score": []}

# Loop over the sets and compute scores
for set_name, data in sets.items():
    for i in range(len(systems)):
        scores2 = mt_scorer(predictions=data["predictions2"][i], references=data["references"])
        scores_dict["System"].append(systems[i])
        scores_dict["Experiment"].append("2")
        scores_dict["Score"].append(scores2["meteor"]["score"])

        scores4 = mt_scorer(predictions=data["predictions4"][i], references=data["references"])
        scores_dict["System"].append(systems[i])
        scores_dict["Experiment"].append("4")
        scores_dict["Score"].append(scores4["meteor"]["score"])

# Convert the dictionary to a DataFrame
scores_df = pd.DataFrame(scores_dict)

# Group by System and Experiment, then calculate the mean
average_scores_df = scores_df.groupby(["System", "Experiment"]).agg(np.mean).reset_index()

# Display the DataFrame with average scores
print(average_scores_df)
````
